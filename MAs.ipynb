```python
btc = pd.read_csv('df_btc.csv')
btc['Datetime'] = pd.to_datetime(btc['Datetime'])

btc.set_index('Datetime', inplace=True)

btc['BTC_MA'] = btc['Close'].rolling(window=20).mean()

btc.tail()

btc.dropna(subset=['BTC_MA'], inplace=True)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(btc['Close'], btc['BTC_MA'], alpha=0.5)
plt.title('Scatterplot between Close and BTC Moving Average')
plt.xlabel('Close Price')
plt.ylabel('BTC 20-Day Moving Average')
plt.grid(True)
plt.show()
```
